```python
import os
import cv2
import numpy as np
from skimage.metrics import structural_similarity as ssim
from google.colab import files

# Function to calculate similarity between two images
def calculate_similarity(image1, image2):
    # Check for empty images
    if image1 is None or image2 is None:
        return 0.0  # Return a low similarity score for empty images

    # Resize images to the same dimensions
    image1_resized = cv2.resize(image1, (image2.shape[1], image2.shape[0]))

    # Convert images to grayscale
    gray_image1 = cv2.cvtColor(image1_resized, cv2.COLOR_BGR2GRAY)
    gray_image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY)

    # Calculate SSIM (Structural Similarity Index)
    similarity_score = ssim(gray_image1, gray_image2)
    return similarity_score

# Define the dataset directory
dataset_path = "/content/drive/MyDrive/medical_images"

# Prompt the user to upload an image
print("Please upload an image for comparison:")
uploaded_files = files.upload()

# Load the user-input image
user_input_image_path = list(uploaded_files.keys())[0]  # Assuming only one image is uploaded
user_input_image = cv2.imread(user_input_image_path)

# Preprocess the user-input image (resize, normalize, etc.)
# (Add your preprocessing code here)

# Iterate through the dataset images
matching_results = []
for disease_folder in os.listdir(dataset_path):
    disease_folder_path = os.path.join(dataset_path, disease_folder)
    for image_file in os.listdir(disease_folder_path):
        image_path = os.path.join(disease_folder_path, image_file)

        # Load dataset image
        dataset_image = cv2.imread(image_path)

        # Preprocess the dataset image (resize, normalize, etc.)
        # (Add your preprocessing code here)

        # Calculate similarity score between user-input image and dataset image
        similarity_score = calculate_similarity(user_input_image, dataset_image)

        matching_results.append((image_path, similarity_score))

# Sort matching results by similarity score
matching_results.sort(key=lambda x: x[1], reverse=True)

# Display top matching results
for i in range(10):  # Display top 10 matching results
    print(f"Matched Image: {matching_results[i][0]}, Similarity Score: {matching_results[i][1]}")
```
